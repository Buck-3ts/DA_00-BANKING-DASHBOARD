```python
bins=[0,100000, 300000, float("inf")]
labels=["low", "medium", "high"]

df["income_band"]=pd.cut(df["Estimated Income"], bins=bins, labels=labels, right=False)
```

```python
numerical_cols =["Estimated Income", "Superannuation Savings", "Credit Card Balance", "Bank Loans", "Bank Deposits", "Checking Accounts", 
                  "Saving Accounts", "Foreign Currency Account", "Business Lending"]
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="Estimated Income", hue='GenderId', kde=True, bins=30)
plt.title('Income Distribution by Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(12,6))
    sns.histplot(data=df,x=col,bins=30,kde=True)
    plt.title(f'{col} distribution')
    plt.tight_layout()
    plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(data=df,y=col)
    plt.title(f'boxplot of {col}')
    plt.show()
```

```python
correlation_matrix=df[numerical_cols].corr()

plt.figure(figsize=(12,6))
sns.heatmap(correlation_matrix, annot=True, cmap='crest',fmt=".2f")
plt.title("heatmap correlation matrix")
plt.show()
```
